```python
import pandas as pd
import os

SOURCE_DIR = r'D:\Major Project\SpamX\ml\dataset\model_dataset'
TARGET_DIR = r'D:\Major Project\SpamX\machine_learning\dataset'

os.makedirs(TARGET_DIR, exist_ok=True)

files = {
    "train": "train1.csv",
    "test": "test.csv",
    "validation": "validation.csv"
}

def transform_to_new_path(filename, save_name):
    input_path = os.path.join(SOURCE_DIR, filename)
    output_path = os.path.join(TARGET_DIR, save_name)
    
    df = pd.read_csv(input_path)
    original_count = len(df)

    df_new = df[['Comment', 'Label']].copy()
    
    df_new['Comment'] = df_new['Comment'].fillna("")

    df_new.to_csv(output_path, index=False)
    
    print(f"Processed {filename}")
    print(f"   - Target: {output_path}")
    print(f"   - Rows:   {len(df_new)} (Exact Match)")

for split, filename in files.items():
    transform_to_new_path(filename, f"{split}.csv")

print(f"\nFiles successfully moved and modified in: {TARGET_DIR}")
```

```python
import pandas as pd
import os
```

```python
CALIB_PATH = r'D:\Major Project\SpamX\machine_learning\dataset\calibration.csv'
TRAIN_PATH = r'D:\Major Project\SpamX\machine_learning\dataset\train.csv'
OUTPUT_PATH = r'D:\Major Project\SpamX\machine_learning\dataset\calibrated.csv'
```

```python
df_missed = pd.read_csv(CALIB_PATH)
df_old = pd.read_csv(TRAIN_PATH)
```

```python
ham_pool = df_old[df_old['Label'] == 0]
spam_pool = df_old[df_old['Label'] == 1]
```

```python
balanced_ham = ham_pool.sample(n=len(df_missed), random_state=42)
```

```python
anchor_ham = ham_pool.sample(n=min(250, len(ham_pool)), random_state=7)
anchor_spam = spam_pool.sample(n=min(250, len(spam_pool)), random_state=7)
```

```python
final_df = pd.concat([df_missed, balanced_ham, anchor_ham, anchor_spam], ignore_index=True)
```

```python
final_df = final_df.drop_duplicates(subset=['Comment'])
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
final_df.to_csv(OUTPUT_PATH, index=False)
```

```python
print(f"SUCCESSFULL: {OUTPUT_PATH}")
print(f"Total Samples: {len(final_df)}")
print(f"Distribution:\n{final_df['Label'].value_counts()}")
```

```python
import pandas as pd
import csv

NEW_DATA = r"D:\Major Project\SpamX\machine_learning\dataset\calibration1.csv"
OLD_CALIB = r"D:\Major Project\SpamX\machine_learning\dataset\calibrated.csv"
OLD_TRAIN = r"D:\Major Project\SpamX\machine_learning\dataset\train.csv"
OUTPUT = r"D:\Major Project\SpamX\machine_learning\dataset\calibrated1.csv"

def prepare_balanced_dataset():
    print("Processing Datasets...")
    
    df_new = pd.read_csv(NEW_DATA, engine='python', on_bad_lines='warn')
    df_old_c = pd.read_csv(OLD_CALIB, engine='python', on_bad_lines='warn')
    df_old_t = pd.read_csv(OLD_TRAIN, engine='python', on_bad_lines='warn')

    combined = pd.concat([df_new, df_old_c, df_old_t], axis=0, ignore_index=True)
    
    initial_count = len(combined)
    combined.drop_duplicates(subset=['Comment'], keep='first', inplace=True)
    
    spam_df = combined[combined['Label'] == 1]
    ham_df = combined[combined['Label'] == 0]
    
    print(f"Initial Balance - Spam: {len(spam_df)}, Ham: {len(ham_df)}")
    
    min_size = min(len(spam_df), len(ham_df))
    balanced_df = pd.concat([
        spam_df.sample(min_size, random_state=42),
        ham_df.sample(min_size, random_state=42)
    ])

    final_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
    final_df.to_csv(OUTPUT, index=False, quoting=csv.QUOTE_ALL, encoding='utf-8')
    
    print("-" * 30)
    print(f"Success!")
    print(f"Final Dataset Size: {len(final_df)}")
    print(f"Duplicates Removed: {initial_count - len(combined)}")
    print(f"Class Balance: 50/50 (Spam: {min_size}, Ham: {min_size})")

if __name__ == "__main__":
    prepare_balanced_dataset()
```
